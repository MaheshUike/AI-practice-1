```python
X = df.drop("Sales", axis =1)
y = df["Sales"]
```

# **Linear Regression class**

lr.LinearRegression()

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

Let's train the model using linear-regression for that we have fit()

* lr.fit(X_train, y_train)  both 80 percent data

```python
lr.fit(X_train, y_train)
```

**Let's create the prediction variable  call it y_pred**

```python
y_pred = lr.predict(X_test) # 20 percent input testing data
```

```python
y_pred
```

**Now model is trained but how can we know whether the model is accurate for this we check accuracy using**

* Metrics R2_score
* import R2_Score and mean_absolute_error()



---

* mean_absolute_error(20%_test_data_which model never saw{y_test}, predicted data by model {y_pred})



---

* r2_score(20%_test_data_which model never seen{y_test}, predicted data by model {y_pred})

```python
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

**Here We have created the prediction model on our own data with 90 percent accurancy**

```python
def lr_prediction(tv_budget, radio_budget, newspaper_budget):
  feature = np.array([[tv_budget, radio_budget, newspaper_budget]]) # here we have created the  2d array as per the model data
  result = lr.predict(feature).reshape(1,-1)
  return result[0]
```

```python
tv_budget = 236.0
radio_budget = 41.1
newspaper_budget = 19.2

predicted_data = lr_prediction(tv_budget,radio_budget, newspaper_budget)
```
