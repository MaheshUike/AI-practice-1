```python
# here in this above line we can see the episode numbers are in side (episode number) so we need to extract the  string and add it into another column

# def extract_text (txt="Bleach: Sennen Kessen-henTV (13 eps)Oct 2022 - Dec 2022474,138 members"):
def extract_text (txt):
  check = False  # this check will become true if we found ( ) else it will remain false
  data = "" # here we have created the  empty string we will concat the  string  here

  for i in txt:
    if i == ")":   # this is condition for if close paranthesis found our search is over  so leave the loop
      break

    if check == True:
      data = data+i # here we are concatinating the empty string with value of i it will become (episode number)

    if i == "(":   # with this if I  = (   means we found our text start point for episode
      check = True

  return data


extract_text()
```

```python
# lets extract all the episode count here first

df["Title"].apply(extract_text)
```

```python
#make a new column for episode count

# lets create a column now and then add the above data

df["Episodes"] = df["Title"].apply(extract_text)

df
```

```python
# now major issue is Epsiodes are in string format we cannot perform aggregation operation on it
# so lets replace eps with empty string

df['Episodes'].str.replace(" eps","")
```

```python
# so we have successfully removed the  eps from the string so now lets add it inside the episode column
df["Episodes"] = df['Episodes'].str.replace(" eps","")
df
```

```python
# let's create a function to extract time stamp

# def extract_time_stamp(txt = "BrotherhoodTV (64 eps)Apr 2009 - Jul 20103,218,472 membersManga StoreVolume 1€4.58Preview" ):
def extract_time_stamp(txt ):
  check = False
  data = ""
# Same as before
  for  i in range(len(txt)):
    if txt[i]==")":
      # print(txt[i])  #so here we are able to find )
      for j in range(i+1, i+20): # with this loop we are checking next 18 charachters after )
        data = data+txt[j]

      return data






extract_time_stamp()
```

```python
df["Title"].apply(extract_time_stamp)
```

```python
# lets just create new column for time stamp now

```

```python
#make a new column for time stamp
df["Total Time"] = df["Title"].apply(extract_time_stamp)
df
```

```python
from dateutil.relativedelta import relativedelta
from datetime import datetime

def calculate_total_months(period):
    try:
        start_str, end_str = period.split(' - ')
        start_date = datetime.strptime(start_str, '%b %Y')
        end_date = datetime.strptime(end_str, '%b %Y')
        r = relativedelta(end_date, start_date)
        return r.years * 12 + r.months + 1  # +1 to include the starting month
    except:
        return None

df['Months'] = df['Total Time'].apply(calculate_total_months)
```
